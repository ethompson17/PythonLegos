# Task #1: What percentage of all licensed sets ever released were Star Wars Themed?

# Task #2: In which year was Star Wars not the most popular licensed theme?

```python
licensed = merged[merged['is_licensed']== True]
```

```python
licensed['count'] = 1
licensed.groupby(['year','parent_theme']).sum()
```

```python
licensed
```

```python
licensed_summed = licensed.groupby(['year','parent_theme']).sum().reset_index()
licensed_summed.head(100)

maxdf = licensed_summed.sort_values('count', ascending = False).drop_duplicates('year')
maxdf.sort_values('year', ascending = False).drop(columns = ['num_parts','id','is_licensed','count'], axis=1)
```

# -- Answer: 2017 --
